```python
import nltk
from nltk.tokenize import word_tokenize, wordpunct_tokenize, TweetTokenizer
from nltk.tokenize.treebank import TreebankWordTokenizer
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
text = "Natural Language Processing is amazing! I love learning NLP."

print("Original Text:", text)
```

```python
# Whitespace tokenization
print("\nWhitespace Tokenization:")
print(text.split())
```

```python
# Punctuation based
print("\nWordPunct Tokenization:")
print(wordpunct_tokenize(text))
```

```python
# Treebank tokenizer
tree = TreebankWordTokenizer()
print("\nTreebank Tokenizer:")
print(tree.tokenize(text))
```

```python
# Tweet tokenizer
tweet = TweetTokenizer()
print("\nTweet Tokenizer:")
print(tweet.tokenize(text))
```

```python
# Stemming
porter = PorterStemmer()
snowball = SnowballStemmer("english")

words = word_tokenize(text)
```

```python
print("\nPorter Stemming:")
print([porter.stem(w) for w in words])
```

```python
print("\nSnowball Stemming:")
print([snowball.stem(w) for w in words])
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()

print("\nLemmatization:")
print([lemmatizer.lemmatize(w) for w in words])
```
